#Combining into functions

```python
def derv_x(x):
  x = torch.tensor(x/1.0, requires_grad=True)
  y = x ** 2
  z = torch.sin(y)
  z.backward()
  return x.grad
```

```python
derv_x(4)
```

#Weights and Biases

```python
#Inputs

x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0)
b = torch.tensor(0.0)
```

```python
def binary_cross_entropy_loss(prediction,target):
  epsilon = 1e-8
  prediction = torch.clamp(prediction, epsilon, 1-epsilon)
  loss = -((target * torch.log(prediction))+((1-target) * torch.log(1-prediction)))
  return loss
```

```python
#Linear Transformation
z = w * x + b
y_pred = torch.sigmoid(z)
```

```python
loss = binary_cross_entropy_loss(y_pred,y)
```

```python
z, y, loss
```

```python
#derivatives
dloss_dy_pred = (y_pred - y) / (y_pred * (1-y_pred))

dy_pred_dz = y_pred * (1 - y_pred)

dz_dw = x

dz_db = 1
```

```python
dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
dL_dw, dL_db
```

#Using autograd

```python
#Inputs

x = torch.tensor(6.7)
y = torch.tensor(0.0)
w = torch.tensor(1.0, requires_grad=True)
b = torch.tensor(0.0, requires_grad=True)
```

```python
z = w * x + b
y_pred = torch.sigmoid(z)
```

```python
loss = binary_cross_entropy_loss(y_pred,y)
```

```python
loss.backward()
```

```python
print(w.grad)
```

```python
print(b.grad)
```
